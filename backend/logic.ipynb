```python
from typing import Optional
from db import get_conn

def choose_best_slot(L: float, W: float, H: float, weight: float) -> Optional[str]:
    conn = get_conn()
    cur = conn.cursor()
    cur.execute("""
      SELECT slot_id, (L-?)+(W-?)+(H-?) AS score
      FROM slots
      WHERE occupied = 0
        AND L >= ? AND W >= ? AND H >= ?
        AND max_weight >= ?
      ORDER BY score ASC
      LIMIT 1;
    """, (L, W, H, L, W, H, weight))
    row = cur.fetchone()
    conn.close()
    return row["slot_id"] if row else None

def set_slot(slot_id: str, occupied: bool, item_id: Optional[str], weight: float):
    conn = get_conn()
    cur = conn.cursor()
    cur.execute("""
      UPDATE slots
      SET occupied = ?, current_item_id = ?, current_weight = ?
      WHERE slot_id = ?;
    """, (1 if occupied else 0, item_id, weight, slot_id))
    conn.commit()
    conn.close()
```
